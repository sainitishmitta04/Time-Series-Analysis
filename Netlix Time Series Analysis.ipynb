# Rolling

```python
df['Open'].head(20)
```

```python
df['Open:10 days rolling'] = df['Open'].rolling(10).mean()
```

```python
df['Open:10 days rolling'].head(20)
```

```python
df[['Open','Open:10 days rolling']].head(20)
```

```python
df[['Open','Open:10 days rolling']].plot(figsize=(15,5))
plt.show()
```

```python
df['Open:20 days rolling'] = df['Open'].rolling(window=20,min_periods=1).mean()
df['Open:50 days rolling'] = df['Open'].rolling(window=50,min_periods=1).mean()
df['Open:100 days rolling'] = df['Open'].rolling(window=100,min_periods=1).mean()
```

```python
df[['Open','Open:10 days rolling','Open:20 days rolling','Open:50 days rolling','Open:50 days rolling']].plot(xlim=['2015-01-01','2024-01-01'],figsize=(15,6))
```

## Resampling

```python
df =  yf.download(tickers = "NFLX")
```

```python
df
```

```python
#year end frequency
df.resample(rule='A').max()
```

```python
df['Adj Close'].resample(rule='3Y').mean().plot(kind='bar',figsize=(10,4))
plt.title('3 Year End Mean Adj Close Price for Netflix')
plt.show()
```

```python
#month end frequency
df.resample(rule='M').mean()
```

```python
#month end frequency
df.resample(rule='M').max()['Open'].plot(figsize=(15,6))
plt.title('Month End Opening Price for Netflix ')
plt.show()
```
